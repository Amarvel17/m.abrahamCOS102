```python
print(f"Welcome to Marvies financial calculator!!")

def display_calculators():
    print("WE HAVE:")
    print("1. Simple interest calculator")
    print("2. Compound interest calculator")
    print("3. Annuity plan calculator")

# Function to calculate simple interest
def calculate_simple_interest(principal, rate, time):
    return principal*(1+(rate/100)*time)

# input for simple interest
def simple_interest_calculator():
    try:
        principal = float(input("Enter the principal: "))
        rate = float(input("Enter the rate: "))
        time = float(input("Enter the time: "))

     # Calculate and display the simple interest
        simple_interest = calculate_simple_interest(principal, rate, time)
        print(f"The simple interest is : {simple_interest} ")
    except ValueError:
        print("Please enter a valid number.")
        simple_interest_calculator()

def calculate_compound_interest(principal, rate, time, n):
    return principal*((1+(rate/n))**(n*time))

# input for compound interest
def compound_interest_calculator():
    try:
        principal = float(input("Enter the principal: "))
        rate = float(input("Enter the rate: "))
        time = float(input("Enter the time: "))
        n = float(input("Enter n: "))

     # Calculate and display the compound interest
        compound_interest = calculate_compound_interest(principal, rate, time, n)
        print(f"The compound interest is : {compound_interest} ")
    except ValueError:
        print("Please enter a valid number.")
        compound_interest_calculator()

def calculate_annuity_plan(rate, time, pmt, n):
    return pmt*((((1+rate/n)**(n*time))-1)/(rate/n))

# input for annuity plan
def annuity_plan_calculator():
    try:
        pmt = float(input("Enter the pmt: "))
        rate = float(input("Enter the rate: "))
        time = float(input("Enter the time: "))
        n= float(input("Enter n: "))

     # Calculate and display the annuity plan
        annuity_plan= calculate_annuity_plan(pmt, rate, time, n)
        print(f"The annuity plan is : {annuity_plan} ")
    except ValueError:
        print("Please enter a valid number.")
        annuity_plan_calculator()

def select_calculator():
    display_calculators()
    
    try:
        choice = int(input("Enter the calculator you would like to use (1, 2 or 3) :"))
        
        if choice==1:
           simple_interest_calculator()
        elif choice==2:
           compound_interest_calculator()
        elif choice==3:
           annuity_plan_calculator()
        else:
           print("Invalid choice! Please select a number between 1 and 3")
           select_calculator()
    except ValueError:
        print("Please enter a valid number.")
        select_calculator()
        

select_calculator()



```
