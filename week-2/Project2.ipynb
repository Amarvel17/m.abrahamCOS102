```python
print(f"Welcome to Marvies arithmetic calculator!!")

def display_calculators():
    print("WE HAVE:")
    print("1. Quartic equation calculator")
    print("2. Cubic equation calculator")
    print("3. Quadratic equation calculator")

import numpy as np
def solve_quartic():
    print("Solve the quartic equation ax^4 + bx^3 + cx^2 + dx + e = 0")

# Take inputs for the coefficients
    a = float(input("Enter the coefficient a (for x^4): "))
    b = float(input("Enter the coefficient b (for x^3): "))
    c = float(input("Enter the coefficient c (for x^2): "))
    d = float(input("Enter the coefficient d (for x): "))
    e = float(input("Enter the constant e: "))

    coefficients = [a, b, c, d, e]

    roots = np.roots(coefficients)

    print("The roots of the quartic equation are:")
    for root in roots:
        print(root)


import numpy as np
def solve_cubic():
    print("Solve the cubic equation ax^3 + bx^2 + cx + d= 0")

    # Take inputs for the coefficients
    a = float(input("Enter the coefficient a (for x^3): "))
    b = float(input("Enter the coefficient b (for x^2): "))
    c = float(input("Enter the coefficient c (for x): "))
    d = float(input("Enter the constant d: "))

    coefficients = [a, b, c, d, e]

    roots = np.roots(coefficients)

    print("The roots of the cubic equation are:")
    for root in roots:
        print(root)


import numpy as np
def solve_quadratic():
    print("Solve the quadratic equation ax^2 + bx + c = 0")

# Take inputs for the coefficients
    a = float(input("Enter the coefficient a (for x^2): "))
    b = float(input("Enter the coefficient b (for x): "))
    c = float(input("Enter the constant c: "))

    coefficients = [a, b, c, d, e]

    roots = np.roots(coefficients)

    print("The roots of the quadratic equation are:")
    for root in roots:
        print(root)

def select_calculator():
    display_calculators()
    
    try:
        choice = int(input("Enter the calculator you would like to use (1, 2 or 3) :"))
        
        if choice==1:
           solve_quartic()
        elif choice==2:
           solve_cubic()
        elif choice==3:
           solve_quadratic()
        else:
           print("Invalid choice! Please select a number between 1 and 3")
           select_calculator()
    except ValueError:
        print("Please enter a valid number.")
        select_calculator()
        

select_calculator()









 
```
